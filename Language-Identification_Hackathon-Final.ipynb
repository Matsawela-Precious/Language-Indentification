## 5. Pre-Processing and Cleaning Data
---

In the pre-processing and cleaning data section we prepare our raw data for further analysis or machine learning tasks. This involves us transforming and cleaning the data to make it suitable for the specific requirements where required.

We remove the noise in the data sets such as punctuation and numbers making the data more usable. The below function aims to remove HTML characters, numbers, newlines, and extra whitespace, while also converting the text to lowercase.

```python
def clean_text(text):
    # replace the html characters with " "
    text=re.sub('<.*?>', ' ', text)
    
    #Removal of numbers
    text = re.sub(r'\d+', ' ', text)
    
    # will replace newline with space
    text = re.sub("\n"," ",text)
    
    # will convert to lower case
    text = text.lower()
    
    # will split and join the words
    text=' '.join(text.split())
    
    return text
```

```python
#we then apply the text processing to our data
df_train['text'] = df_train['text'].apply(clean_text)
df_test['text'] = df_test['text'].apply(clean_text)
```

By applying the clean_text function to the 'text' column of both datasets (test and train), the code performs the same text cleaning operations on the text data in both training and testing sets. This helps ensure consistency in the preprocessing steps applied to the text before further analysis or modeling.

```python
# Replace '.txt' with 'text file'
df_train['text'] = df_train['text'].str.replace(".txt", " text file")
df_test['text'] = df_test['text'].str.replace(".txt", " text file")
```

The code used helps us replace '.txt' substring with the more descriptive phrase 'text file'.

## 6. Modeling
---

In this section we focus on training different models so that we are able to select a suitable model to make predictions or derive insights from our data. 


```python
#we devide our features into the x and y as follows
X = df_train['text']
y = df_train['lang_id']
```

The 'text' column of the train dataset is assigned to X and the 'lang_id' column of thr train data is assigned to y. This separation of features into X (input) and y (output) is commonly done in machine learning tasks to prepare the data for model training. We can then use X as the input data and y as the corresponding labels for training our machine learning model.

```python
vector = TfidfVectorizer(sublinear_tf=True)
vector = TfidfVectorizer(ngram_range=(1,2), min_df=2)
X_vector = vector.fit_transform(X)
```

This method fits the vectorizer on the data and transforms the input text into a matrix representation using the TF-IDF (Term Frequency-Inverse Document Frequency) weighting scheme. The resulting X_vector variable represents the transformed text data in a numerical format suitable for training machine learning models.

```python
X_vector.shape
```

```python
# Splitting the previously defined features and label of your dataset
X_train, X_test, y_train, y_test = train_test_split(X_vector, y, test_size=0.2, random_state=42)
```

The provided code splits the previously defined features (X_vector) and labels (y) of the dataset into training and testing sets using the train_test_split function. These sets can then be used for training and evaluating machine learning models.

**BUILDING CLASSIFIERS.**

In this subsection we build our classifiers .

```python
names = ['Logistic Regression', 'Nearest Neighbors', 
         'Naive Bayes']
```

```python
classifiers = [
    LogisticRegression(), 
    KNeighborsClassifier(3),
    MultinomialNB(alpha=0.3)
]
```

```python
results = []

models = {}
confusion = {}
class_report = {}


for name, clf in zip(names, classifiers):    
    print ('Fitting {:s} model...'.format(name))
    run_time = %timeit -q -o clf.fit(X_train, y_train)
    
    print ('... predicting')
    y_pred = clf.predict(X_train)   
    y_pred_test = clf.predict(X_test)
    
    print ('... scoring')
    accuracy  = metrics.accuracy_score(y_train, y_pred)
    precision = metrics.precision_score(y_train, y_pred, average = 'weighted')
    recall    = metrics.recall_score(y_train, y_pred, average = 'weighted')
    
    f1        = metrics.f1_score(y_train, y_pred, average = 'weighted')    
    f1_test   = metrics.f1_score(y_test, y_pred_test, average = 'weighted')    
    
    # Save the results to dictionaries
    models[name] = clf    
    confusion[name] = metrics.confusion_matrix(y_train, y_pred)
    class_report[name] = metrics.classification_report(y_train, y_pred)
    
    results.append([name, accuracy, precision, recall, f1, f1_test, run_time.best])

    
results = pd.DataFrame(results, columns=['Classifier', 'Accuracy', 'Precision', 'Recall', 'F1 Train', 'F1 Test', 'Train Time'])
results.set_index('Classifier', inplace= True)

print ('... All done!')
```
